```python
from sys import path
path.append('..')

from knn_sampler import KNNSampler
from sklearn.datasets import load_wine
```

```python
data = load_wine(as_frame=True)['data']
```

```python
data.sample(10)
```

```python
sampler = KNNSampler()
```

```python
sampler.fit(data,['malic_acid','ash'], 'hue', feature_weights = [1,2], scaler = 'RobustScaler')
```

```python
f = lambda x: x #samples proportional to distance
sampler.sample(data.sample(10), sampling_weights = f)
```
